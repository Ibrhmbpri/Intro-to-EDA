```python
# To find out how many coasters with the same name, in the same location, and same opening date appear more than once in the dataset.
df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date']).sum()
```

```python
# To create a cleaned version of the DataFrame where: Duplicate coasters (based on name, location, and date) are removed, The row index is reset to be tidy and continuous
# See the output of the previous command without .reset_index(drop=True)
# The backslash (\) in that line is used to break a long line of code into multiple lines for better readability.
# .copy() is used to make a true, independent copy of the DataFrame, so that future changes won’t affect the original data (or raise warnings).

df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])]\
     .reset_index(drop=True).copy()
```
